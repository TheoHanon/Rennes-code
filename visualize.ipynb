# Weather Data

```python
weather_ds = np.load("data/weather/20260108_00z/weather_test.npz")

vals = weather_ds["dpt2m"]
times = weather_ds["time"]
```

```python
fig = plt.figure()
im = plt.imshow(np.flipud(vals[0]), extent = [0, 2*np.pi, -np.pi/2, np.pi/2], cmap = "RdBu")
plt.colorbar(im)

def update(frame):
    im.set_array(np.flipud(vals[frame]))
    plt.title(f"Time: {times[frame]:.2f}")
    return im,


ani = FuncAnimation(fig,update,frames = len(times))
from IPython.display import HTML
HTML(ani.to_html5_video())
```
